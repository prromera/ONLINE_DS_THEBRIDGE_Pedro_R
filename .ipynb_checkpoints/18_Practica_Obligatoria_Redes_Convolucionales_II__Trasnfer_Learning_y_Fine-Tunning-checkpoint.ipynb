```python
def read_data(directorio, reshape_dim = (75,75)):
    X = [] 
    y = []
    for folder in os.listdir(directorio):
        print(folder)
        if os.path.isdir('/'.join([directorio, folder])):
            for file in os.listdir('/'.join([directorio, folder])):

                image = imread('/'.join([directorio, folder, file]))
                image = cv2.resize(image, reshape_dim) # Redimensionamos las imágenes a 32x32

                X.append(image)
                y.append(folder)

    return np.array(X),np.array(y)
```

```python
def show_images_batch(paisajes, names = [], n_cols = 5, size_scale = 2):
    n_rows = ((len(paisajes) - 1) // n_cols + 1)
    plt.figure(figsize=(n_cols * size_scale, n_rows * 1.1*size_scale))
    for index, paisaje in enumerate(paisajes):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(paisaje, cmap = "Greys")
        plt.axis("off")
        if len(names):
            plt.title(names[index])
```

```python
ROOT_PATH = './data/'
TRAIN_PATH = ROOT_PATH + "seg_train/"
TEST_PATH = ROOT_PATH + "seg_test/"
```

```python
X_train,y_train = read_data(TRAIN_PATH)
X_test, y_test = read_data(TEST_PATH)
```

```python
print('X train shape: ', X_train.shape)
print('Y train shape: ', y_train.shape)
```

```python
indices = np.random.randint(0,len(X_train),20)
show_images_batch(X_train[indices], names = y_train[indices], n_cols = 5)
```

```python
targets = pd.Series(y_train)
targets.value_counts(True)
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
# Barajar los datos
X_train, y_train = shuffle(X_train, y_train)
X_test, y_test = shuffle(X_test, y_test)
```

```python
mapa = {tipo:indice for indice,tipo in enumerate(targets.unique())} # Rehusando la serie que creamos para ver la distribución del target
mapa
```

```python
y_train_num = np.array([mapa[y] for y in y_train])
y_test_num = np.array([mapa[y] for y in y_test])
```

```python
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(75,75, 3))
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(6, activation='softmax')(x)
```

```python
model_transfer_learning = Model(inputs=base_model.input, outputs=predictions)
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
model_transfer_learning.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model_transfer_learning.summary()
```

```python
model_transfer_learning.fit(X_train, y_train_num, epochs=10, validation_data=(X_test, y_test_num))
```

```python
history_df = pd.DataFrame(model_transfer_learning.history)
plt.plot(history_df['accuracy'], label = 'Accuracy')
plt.plot(history_df['val_accuracy'], label = 'Val_accuracy')

plt.legend();
```

```python
test_loss, test_accuracy = model_transfer_learning.evaluate(X_test, y_test_num)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
y_pred = np.argmax(model_transfer_learning.predict(X_test), axis=-1)
```

```python
print(classification_report(y_test_num, y_pred))
```
